```python
import time
import pandas as pd
from selenium import webdriver
```

```python
chrome_options = webdriver.ChromeOptions()
chrome_options.add_argument('--headless')   #active headless mode
driver = webdriver.Chrome(options=chrome_options, executable_path='C:/chromedriver.exe')
```

```python
driver.get('https://www.cwb.gov.tw/V8/E/W/OBS_Map.html')  #go to the CWB website
time.sleep(2)
```

```python
#date and time   
date = driver.find_element_by_id("preestimate").text
    
#city
citys = []
WebElements_city = driver.find_elements_by_xpath("//span[@class='city']")
for city in WebElements_city:
    if city.text != '':
        citys.append(city.text)
        
#temperature
temps = []
WebElements_temp = driver.find_elements_by_xpath("//span/i")
for temp in WebElements_temp:
    if temp.text != '':
        temps.append(temp.text)
```

```python
#humidity
element = driver.find_element_by_xpath("//li/a[@id='nowhum']")
driver.execute_script("arguments[0].click();", element)
time.sleep(1)
    
hums = []
WebElements_hum = driver.find_elements_by_xpath("//span/i")
for hum in WebElements_hum:
    if hum.text != '':
        hums.append(hum.text)
    

#precipitation
element = driver.find_element_by_xpath("//li/a[@id='nowrain']")
driver.execute_script("arguments[0].click();", element)
time.sleep(1)
    
rains = []
WebElements_rain = driver.find_elements_by_xpath("//span/i")
for rain in WebElements_rain:
    if rain.text != '':
        rains.append(rain.text)

driver.close()
```

```python
data = {'City': citys,
        'Temperature': temps, 
        'Humidity': hums, 
        'Precipitation': rains} 
df = pd.DataFrame(data) 
df
```

```python
#turn dataframe to csv file
df.to_csv('D:\TempHumRain.csv', index=False)

#add date and time 
with open('D:\TempHumRain.csv', 'a') as file:
    file.write('\n%s\n' %date)
file.close()
```
